通过实验练习，你可能发现**人脸年龄检测、性别检测**的结果有时候很离谱。这是因为每次检测的时候，我们是针对整张图进行的预测，图像中除了人脸以外，还有大量的非人脸部分。

项目文件中提供了一个综合性的人脸检测函数`predict_age_and_gender(image)`，这个函数的特点如下：
- 首先，检测图像中的人脸
- 其次，如果图像中存在人脸，那么将人脸区域裁剪出来，得到仅有人脸的小图像
- 然后，对小图像进行人脸性别、年龄的检测
- 函数还可以将人脸区域以及人脸性别、年龄的检测结果显示在图像上
- 使用这个函数，将同时得到性别、年龄两个返回值，可以尝试使用两个变量同时接收他们。

### 自主实践-2：

```python
gender, age = predict_age_and_gender(img) # 获取指定图像中人的性别，年龄，使用变量gender、age接收

print(f"图像中人脸的性别为：{gender}") # 输出gender变量的值
print(f"图像中人脸的年龄段为：{age}") # 输出age变量的值
```
